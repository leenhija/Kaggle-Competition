```python
stroke= pd.read_csv('/content/train (2).csv')
stroke
```

```python
stroke.info()
```

```python
stroke.describe()
```

```python
stroke.isna().sum()
```

```python
stroke.dropna(inplace=True)
```

```python
stroke.isna().sum()
```

```python
stroke.head()
```

```python
id = stroke['id']
```

```python
stroke.drop('id',axis=1, inplace = True)
```

```python
gender = pd.get_dummies(stroke['gender'],drop_first=True)
gender = gender.drop('Other', axis=1)
gender
```

```python
ever_married = pd.get_dummies(stroke['ever_married'],drop_first=True) 
ever_married
```

```python
work_type  = pd.get_dummies(stroke['work_type'])
work_type 
```

```python
Residence_type  = pd.get_dummies(stroke['Residence_type'])
Residence_type 
```

```python
smoking_status = pd.get_dummies(stroke['smoking_status'])
smoking_status
```

```python
stroke=pd.concat([gender,stroke['age'],stroke['hypertension'],stroke['heart_disease'],ever_married,work_type,Residence_type,stroke['avg_glucose_level'],stroke['bmi'],smoking_status,stroke['stroke']],axis=1)
stroke
```

```python
stroke.rename(columns={'Male':'Gender','Yes':'ever_married','Unknown':'Unknown_Smoking_Status'},inplace=True)
stroke
```

```python
X = stroke.drop('stroke', axis=1)
y= stroke['stroke']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn import metrics 
```

```python
metrics.accuracy_score(y_test,y_predict)
```
